```python
df_1960 = pd.DataFrame({
    'Life Expectancy': df1["1960"],
    'Fertility Rate': df3["1960"]
})
df_1960 = df_1960.dropna()
plt.scatter(df_1960["Life Expectancy"], df_1960["Fertility Rate"])
plt.xlabel('Life Expectancy')
plt.ylabel('Fertility Rate')
plt.xlim(25, 85)
plt.ylim(1, 9)
```

```python
def draw_plot(year):
    new_df = pd.DataFrame({
        'Life Expectancy': df1[year],
        'Fertility Rate': df3[year]
    })
    new_df = new_df.dropna()
    
    plt.scatter(new_df["Life Expectancy"], new_df["Fertility Rate"])
    plt.xlabel('Life Expectancy')
    plt.ylabel('Fertility Rate')
    plt.xlim(20, 90)
    plt.ylim(0, 10)
    plot = plt.gcf()
    return plot
```

```python
draw_plot('1990')
```

```python
for i in range(1960, 2010):
    draw_plot(str(i))
```

```python
for i in range(1960, 2010):
    plot = draw_plot(str(i))
    plot.savefig('scatterplot_' + str(i) + '.png', dpi=300)
    plot.close()
```
