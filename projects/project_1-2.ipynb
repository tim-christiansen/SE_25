### Task 2
Read all files yob1880.txt, yob1881.txt ... yob2021.txt. Add an extra column for the year. Concatenate them into a single data structure.


```python
import glob
import re
```

```python
all_files = glob.glob("./names/*.txt")
all_files
```

```python
li = []

for filename in all_files:
    df = pd.read_csv(
        filename,
        sep=',',
        names=['Name','Gender','Amount'], #set columns names
        header=None
    )
    df['Year'] = re.findall(r'\d+', filename)[0]
    li.append(df)


frame = pd.concat(li, axis=0, ignore_index=True)

data = list(round((frame['Amount'] / frame['Amount'].sum()*100),8)) # creates a list of 
frame['Percentage'] = data

years = frame['Year'].unique()
frame
```
