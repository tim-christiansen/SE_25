```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset('penguins')
```

```python
df.head()
```

## How does a typical penguin beak look like?

```python
sns.scatterplot(data=df, x='bill_length_mm', y='bill_depth_mm', hue='species')
```

```python
species = df.set_index('species')  # new DF with different index
```

Most basic typical indicator is the average. To be more spefic I'm grouping by species

```python
adelie = species.loc["Adelie"]
chinstrap = species.loc["Chinstrap"]
gentoo = species.loc["Gentoo"]
```

```python
adelie["bill_length_mm"].mean()
```

```python
adelie["bill_depth_mm"].mean()
```

```python
chinstrap["bill_length_mm"].mean()
```

```python
chinstrap["bill_depth_mm"].mean()
```

```python
gentoo["bill_length_mm"].mean()
```

```python
gentoo["bill_depth_mm"].mean()
```
