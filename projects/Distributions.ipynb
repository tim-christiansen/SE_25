# Distributions

* gaussian == normal
* uniform
* Poisson
* binomial
* exponential

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

### Normal or Gaussian

```python
np.random.seed(42)  # <-- see Douglas Adams
n = np.random.normal(100, 10, size=10_000_000)
pd.DataFrame(n).hist(bins=100)
```

### Uniform distribution
rolling a dice

```python
a = np.random.uniform(0, 10, size=10_000_000)
pd.DataFrame(a).hist(bins=100)
```

### Binomial
coin toss

```python
b = np.random.binomial(n=500, p=0.5, size=100000)  # n: number of coin tosses, p: probability of the coin
pd.DataFrame(b).hist(bins=25)
```

### Poisson

number of times a rare event occurs (e.g. 2 penguins dating)

```python
L = 60/11  # ~5.5 penguin dating events per hour ("1 date every 11 minutes")
p = np.random.poisson(lam=L, size=100000)
pd.DataFrame(p).hist(bins=100)
plt.xlabel("events per hour")
```

### Why do we always see the normal distribution?

```python
# dice roll
a = np.random.randint(1, 7, size=100)
a
```

```python
# we roll two dice now and add them together
a = np.random.randint(1, 7, size=(10000, 2))
pd.DataFrame(a.sum(axis=1)).hist(bins=100)
```

```python
# more dice and add them together
a = np.random.randint(1, 7, size=(100000, 10))
pd.DataFrame(a.sum(axis=1)).hist(bins=100)
```

### Central Limit Theorem
if you add up a big number of any distribution, you will end up with a normal distribution.
